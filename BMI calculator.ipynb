```python
import pandas as pd
```

```python
file_csv=pd.read_csv('/Users/krishnamenon/Downloads/hw.txt',header=None)
```

```python
file_csv
```

```python
column_names=['Height_cm','Weight_lbs','Height_m','Weight_kg','BMI','BMI_Categories']
```

```python
data_df=pd.DataFrame(columns=column_names)
```

```python
data_df
```

```python
len_1=int(file_csv.shape[0]/2)
```

```python
len_1
```

```python
for i in range(0,file_csv.shape[0],2):
    #data_df['Height_cm'][i]=file_csv[i]
    data_df.at[i, 'Height_cm'] = file_csv[0][i]
    data_df.at[i, 'Weight_lbs'] = file_csv[0][i+1]
    
    
```

```python
data_df
```

```python
data_df.reset_index(inplace=True)
```

```python
data_df
```

```python
data_df = data_df.drop(columns=['index'])
```

```python
data_df
```

```python
for i in range(0,data_df.shape[0]):
    
    data_df.at[i, 'Height_m'] = (data_df['Height_cm'][i])/100
    data_df.at[i, 'Weight_kg']= (data_df['Weight_lbs'][i])/2.2
```

```python
data_df
```

```python
for i in range(0,data_df.shape[0]):
    data_df.at[i, 'BMI'] = ((data_df['Weight_kg'][i])/(data_df['Height_m'][i]*data_df['Height_m'][i]))
```

```python
data_df
```

```python
for i in range(0,data_df.shape[0]):
    if data_df['BMI'][i]<18.5:
        data_df.at[i, 'BMI_Categories']='Underweight'
    elif (data_df['BMI'][i]>18.5 and data_df['BMI'][i]<24.9):
        data_df.at[i, 'BMI_Categories']='Normal weight'
    elif (data_df['BMI'][i]>25 and data_df['BMI'][i]<29.9):
        data_df.at[i, 'BMI_Categories']='Overweight'
    else:
        data_df.at[i, 'BMI_Categories']='Obesity'
```
